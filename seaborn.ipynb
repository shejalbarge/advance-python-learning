```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips)
```

```python
plt.scatter(tips['total_bill'], tips['tip'])
plt.title("totall bill and tip graph")

plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y='tip', data= tips, color='red')
plt.show()
```

```python
sns.lineplot(x = 'total_bill', y='tip', data= tips, color='red')
plt.show()
```

```python
sns.lineplot(x = 'sex', y='smoker', data= tips, color='red')
plt.show()
```

```python
sns.lineplot(x = 'day', y='time', data= tips, color='red')
plt.show()
```

```python
sns.lineplot(x ='total_bill', y='tip', hue = 'smoker', style='smoker', data= tips, palette = 'Set2', marker='o')
```

```python
plt.hist(tips['tip'], bins=10, color='pink')
plt.show()
```

```python
sns.histplot(data=tips, x='total_bill', color='red', kde=True)
plt.show()
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
corr = tips.corr(numeric_only=True)                 #corr------corr is to store analysis data
sns.heatmap(corr, annot=True, cmap='coolwarm')       #annot--- line box sturcture, cmap= colour type,
plt.show()
```
