```python
import matplotlib.pyplot as plt
```

```python
a = [1, 2, 3, 4, 5]
b = [6, 9, 0, 6, 7]

plt.plot(a, b, linestyle = '--', c olor ="brown")    # plot is keyword used to describe line chart and a drawn on x axis and b drwan on y axis
plt.title("single line chart")      #plt.title used to give title to chart
plt.xlabel("size")                 #plt.xlabel used to give title to x axis
plt.ylabel("colour code")         #plt.ylabel used to give title to y axis

plt.show()
```

```python
sizes = [10, 15, 20, 25, 30]
labels = [ 'good', 'bad', 'poor', 'classy', 'best']
colors = ['red', 'green', 'yellow', 'pink', 'purple']

plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", startangle=140)
plt.title("pie chart")
plt.show()
```
