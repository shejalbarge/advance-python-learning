```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd
df = pd.read_csv(r"C:\Users\sheja\Downloads\sales_data_500 (1).csv")
print(df)
```

```python
sns.lineplot(x='age', y='price', data=df, color='red')
plt.show()
```

```python
sns.scatterplot(x='total_purchase', y='quantity', data=df, color='purple')
plt.show()
```

```python
sns.histplot(x='total_purchase', bins=10, data=df, color='pink', kde=True)
plt.show()
```

```python
sns.violinplot(x='age', y='total_purchase', data=df)
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
